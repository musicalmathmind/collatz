```python
import math
import numpy as np
from sympy import factorint
from fractions import Fraction
def collatz(n):
    orbit = [n]
    d_orbit = None
    while n != 1:
        n = n // 2 if n % 2 == 0 else 3 * n + 1
        if (d_orbit is None and n <= orbit[0]):
            d_orbit = orbit.copy()
        orbit.append(n)
    return orbit, d_orbit

class CollatzInfo():
    def __init__(self, n, universe, n_map, n_count):
        self.n = n
        self.orbit, self.d_orbit = collatz(n)
        self.w = self.d_orbit[-1] // 2
        self.x = n - self.w
        self.a = self.w**2 - self.x**2
        self.b = 2 * self.w * self.x
        self.c = self.w**2 + self.x**2
        self.gcd = gcd_of_three(self.a, self.b, self.c)
        self.y = self.a // n
        self.z = self.b // self.w
        self.d_set = len(self.d_orbit)
        self.d_modulus = n_map[self.d_set]
        self.d_index = n_count[self.d_set]
        self.odd_count = len([x for x in self.d_orbit if x % 2 == 1])
        self.even_count = len([x for x in self.d_orbit if x % 2 == 0])
        n_map[self.d_set] += 1
        n_count[self.d_set] += 1
        self.odd_links = []
        self.M = np.matrix([[self.w, self.x], [self.y, self.z]])
        for i in range(0, 100):
            check = (((2 * 2**i * self.n) - 1) % 3 == 0)
            if (check):
                odd = ((2 * 2**i * self.n) - 1) // 3
                self.odd_links.append(odd)
                if (len(self.odd_links) == 5):
                    break
        if (n_map[self.d_set] >= universe[self.d_set]):
            n_map[self.d_set] = 0


def gen_dset(n_elements):
    elements = []
    for n in range(n_elements):
        elements.append(math.floor(1 + n + n*math.log(3)/math.log(2)))
    return elements

def gen_modulus(nth_element):
    # Increase limit based on requested nth element
    limit = nth_element * 10
    x = [0] * (limit + 1)
    y = [0] * (limit + 1)
    x[1] = 1
    
    sequence = []
    for b in range(2, limit + 1):
        for c in range(2, b + 2):
            y[c] = x[c] + x[c - 1]
        for c in range(2, b + 2):
            x[c] = y[c]
            
        a_n = 0
        for c in range(1, b + 2):
            if ((b + 1 - c) * math.log(3) < b * math.log(2)):
                a_n += x[c]
                x[c] = 0  # Reset x[c] after adding to a_n
                
        if a_n != 0:
            sequence.append(a_n)
            if len(sequence) == nth_element:  # Exit loop once we reach nth element
                break
    
    return sequence


def gcd(a, b):
    """Compute the GCD of two numbers."""
    while b:
        a, b = b, a % b
    return a

def gcd_of_three(a, b, c):
    """Compute the GCD of three numbers."""
    # First find GCD of a and b
    gcd_ab = gcd(a, b)
    # Then find GCD of gcd_ab and c
    return gcd(gcd_ab, c)

def numberToBase(n, b):
    if n == 0:
        return [0]
    digits = []
    while n:
        digits.append(int(n % b))
        n //= b
    return digits[::-1]

import matplotlib.pyplot as plt

def plot_2d_points(points):
    # Extract the x and y coordinates from the points list
    x_coords = [point[0] for point in points]
    y_coords = [point[1] for point in points]
    
    # Create a scatter plot
    plt.figure(figsize=(8, 6))
    plt.scatter(x_coords, y_coords, color='blue', marker='o')
    
    # Add titles and labels
    plt.title('Scatter Plot of 2D Points')
    plt.xlabel('X coordinate')
    plt.ylabel('Y coordinate')
    
    # Show grid
    plt.grid(True)
    
    # Display the plot
    plt.show()


def bin_x(n):
    return bin(n)[2:]


import matplotlib.pyplot as plt
import numpy as np

def draw_unit_circle_with_points(points_groups):
    """
    Draws the unit circle and plots groups of points on it, with each group having a distinct color.

    :param points_groups: A list of lists of points, where each point is a tuple (x, y).
                          Each sublist represents a group of points to be plotted in a distinct color.
    """
    # Create a figure and a subplot
    fig, ax = plt.subplots()

    # Generate points for the unit circle
    circle = plt.Circle((0, 0), 1, color='blue', fill=False)

    # Add the circle to the plot
    ax.add_artist(circle)

    # Set the aspect of the plot to be equal, to maintain the circle shape
    ax.set_aspect('equal', 'box')

    # Set limits to slightly larger than the unit circle to ensure it's fully visible
    ax.set_xlim(-1.1, 1.1)
    ax.set_ylim(-1.1, 1.1)

    # Plot each group of points with a different color
    colors = ['red', 'green', 'yellow', 'purple', 'orange', 'cyan', 'magenta']
    for i, group in enumerate(points_groups):
        # Extract x and y coordinates
        xs, ys = zip(*group)
        # Plot points
        ax.plot(xs, ys, 'o', color=colors[i % len(colors)])

    plt.grid(True)
    plt.title('Unit Circle with Points')
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.show()

def gcd(a, b):
    """Compute the Greatest Common Divisor of a and b."""
    while b:
        a, b = b, a % b
    return a

def generate_primitive_pythagorean_triples(n):
    """Generate the first n primitive Pythagorean triples."""
    triples = []
    m = 2  # Starting value of m
    while len(triples) < n:
        for k in range(1, m):
            # Ensure that m and k are coprime and not both odd
            if (gcd(m, k) == 1) and not (m % 2 == 1 and k % 2 == 1):
                a = m**2 - k**2
                b = 2*m*k
                c = m**2 + k**2
                # Ensure a < b for consistency
                if a > b:
                    a, b = b, a
                triples.append((a, b, c))
                if len(triples) == n:
                    break
        m += 1
    return triples

def prime_factors(n):
    """Return the prime factors of an integer n as a flat array."""
    from sympy import factorint
    
    # Get the prime factors and their multiplicity
    factor_dict = factorint(n)
    factors = []
    
    for factor, multiplicity in factor_dict.items():
        factors.extend([factor] * multiplicity)
    
    return factors

import math

def arc_length_on_unit_circle(point1, point2):
    """
    Calculate the arc length distance between two points on the unit circle.
    
    Parameters:
    - point1: A tuple (x1, y1) representing the first point on the unit circle.
    - point2: A tuple (x2, y2) representing the second point on the unit circle.
    
    Returns:
    - The arc length distance between the two points on the unit circle.
    """
    
    # Extract coordinates
    x1, y1 = point1
    x2, y2 = point2
    
    # Calculate the angle between the points using the dot product formula for unit vectors
    # dot_product = x1*x2 + y1*y2 = cos(theta), where theta is the angle between the vectors in radians
    dot_product = x1*x2 + y1*y2
    angle = math.acos(dot_product)
    
    # Arc length on a unit circle is given by the angle in radians (theta)
    arc_length = angle
    
    return arc_length

# Example usage with points (1, 0) and (0, 1) on the unit circle
arc_length_on_unit_circle((1, 0), (0, 1))

def agv_squares(n):
    squares = [n**2 for n in range(1, n+1)]
    return sum(squares) / n

import numpy as np
from scipy.linalg import schur, eigvals

def are_similar(A, B):
    # Check if matrices are square and of the same size
    if A.shape != B.shape or A.shape[0] != A.shape[1]:
        return False

    # Eigenvalues must be the same (including multiplicities)
    eigenvalues_A = np.sort(eigvals(A))
    eigenvalues_B = np.sort(eigvals(B))
    if not np.allclose(eigenvalues_A, eigenvalues_B):
        return False

    # Use the Schur decomposition to check for similarity in a stable manner
    T_A, _ = schur(A, sort='iuc')
    T_B, _ = schur(B, sort='iuc')
    return np.allclose(T_A, T_B)

arr_1033= np.array([
    [272, 138, 341, 131, 151],
    [366, 199, 130, 320, 18],
    [226, 245, 91, 245, 226],
    [18, 320, 130, 199, 366],
    [151, 131, 341, 138, 272]
])

arr_3301 = np.array([
    [434, 1311, 312,  278,  966],
    [204,  812, 934,  280, 1071],
    [626,  620, 809,  620,  626],
    [1071, 280, 934,  812,  204],
    [966,  278, 312, 1311,  434]
])                   

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_3d_points(points):
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    # Extract x, y, and z coordinates from the points list
    x_coords = [point[0] for point in points]
    y_coords = [point[1] for point in points]
    z_coords = [point[2] for point in points]
    
    # Create a scatter plot
    scatter = ax.scatter(x_coords, y_coords, z_coords, color='blue', marker='o')
    
    # Add titles and labels
    ax.set_title('3D Scatter Plot of Points')
    ax.set_xlabel('X coordinate')
    ax.set_ylabel('Y coordinate')
    ax.set_zlabel('Z coordinate')
    
    # Setting the rotation angle for better visualization
    ax.view_init(elev=20., azim=-35)

    # Show plot
    plt.show()


import plotly.graph_objects as go

def plot_3d_points_interactive(points):
    # Unpack the list of points
    x_coords = [p[0] for p in points]
    y_coords = [p[1] for p in points]
    z_coords = [p[2] for p in points]
    
    # Create a 3D scatter plot
    fig = go.Figure(data=[go.Scatter3d(
        x=x_coords,
        y=y_coords,
        z=z_coords,
        mode='markers',
        marker=dict(
            size=5,
            color='blue',  # Set marker color to blue
            opacity=0.8
        )
    )])
    
    # Update plot layout
    fig.update_layout(
        title='3D Scatter Plot of Points',
        scene=dict(
            xaxis_title='X Axis',
            yaxis_title='Y Axis',
            zaxis_title='Z Axis'
        ),
        margin=dict(l=0, r=0, b=0, t=0)  # Tighten the margin to use more space for the plot
    )
    
    # Show the plot
    fig.show()

```
